```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
def plot_patches_on_image(patch_coords, image_size, max_plot_size=1920):
    # Compute the ratio between the max plot size and the image size
    scale_factor = min(max_plot_size/image_size[0], max_plot_size/image_size[1])
    plot_size = (image_size[0]*scale_factor, image_size[1]*scale_factor)

    # Scale the patch coordinates

    # Create a new figure with a size of max_plot_size
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(2.1*plot_size[0]/100, plot_size[1]/100))

    # Load the image
    image = plt.imread('./Normal_083.tif')

    # Show the image
    ax1.imshow(image, extent=[0, image_size[0], image_size[1], 0])
    ax1.axis('off')

    ax2.imshow(image, extent=[0, image_size[0], image_size[1], 0])
    ax2.axis('off')
    
    # Loop over each patch and plot it on the image
    for coord in patch_coords:
        x, y = coord
        ax2.plot(x, y, 'bs', markersize=3)

    # Set the x and y limits of the plot to the size of the image
    ax1.set_xlim([0, image_size[0]])
    ax1.set_ylim([0, image_size[1]])
    ax2.set_xlim([0, image_size[0]])
    ax2.set_ylim([0, image_size[1]])

    # Show the plot
    plt.show()
```
